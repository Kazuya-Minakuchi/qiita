# アタポン
* アタポンのデータに絞る
* イベント期間(h)を使って、ローカルレベル+時変係数モデルにしてみる

## インポート

```python
import numpy as np
import pandas as pd
import pickle
import datetime
from datetime import timedelta
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font=['IPAMincho'])

import pystan
```

## イベントデータ読み込み

```python
input_dir = '../event_data/'
```

```python
df_all = pd.read_pickle(input_dir + 'b1_output_df.pkl')
df_all
```

```python
# 目的変数が欠損してない行のみ取り出す
df = df_all[~(df_all['2001位'].isnull())].reset_index(drop=True)
df
```

## stanデータ準備

```python
# 欠損でない位置のindex番号+1
# stanは0ではなく1スタートなので+1してる
obs_no = list(map(lambda x: x+1, df_all[(~(df_all['2001位'].isnull()))].index.tolist()))
```

```python
# データ（辞書型）
dat = {
    'T':       len(df_all),         # 全日付の日数
    'len_obs': len(df),               # データがある日付の日数
    'ev_len':  df['期間'].tolist(),   # イベント期間(h)
    'y':       df['2001位'].tolist(), # 観測値
    'obs_no':  obs_no,                # 観測値がある場所（あったらTrue、欠損だとFalse）
}
```

```python
# Stanコード
model = """
data {
  int T;                   // データ取得期間の長さ
  int len_obs;             // 観測値が得られた個数
  vector[len_obs] ev_len;  // イベント期間(h)
  vector[len_obs] y;       // 観測値
  int obs_no[len_obs];     // 観測値が得られた時点
}
parameters {
  vector[len_obs] b_ev_len; // ev_lenの係数
  vector[T] mu;             // 水準成分の推定値
  real<lower=0> s_t;        // ev_lenの係数の変化を表す標準偏差
  real<lower=0> s_w;        // 水準成分の変動の大きさを表す標準偏差
  real<lower=0> s_v;        // 観測誤差の標準偏差
}
transformed parameters {
  vector[len_obs] alpha;
  for(i in 1:len_obs){
    alpha[i] = mu[obs_no[i]] + b_ev_len[i] * ev_len[i];
  }
}
model {
  for(i in 2:T){
    mu[i] ~ normal(mu[i-1], s_w);
  }
  for(i in 2:len_obs){
    b_ev_len[i] ~ normal(b_ev_len[i-1], s_t);
  }
  for(i in 1:len_obs){
    y[i] ~ normal(alpha[i], s_v);
  }
}
"""
```

# Stan実行

```python
# コンパイル
%time stm = pystan.StanModel(model_code=model)
```

```python
# パラメータ設定
n_itr = 20000
n_warmup = n_itr - 1000
chains = 3
```

```python
# 計算
%time fit = stm.sampling(data=dat, iter=n_itr, chains=chains, n_jobs=-1, warmup=n_warmup, algorithm="NUTS", verbose=False)
```

```python
# 時間かかるのでfit後のファイル保存
save_dir = '../event_data/'
with open(save_dir + 'b4_stan.pickle', mode="wb") as f:
    pickle.dump(fit, f)
```

```python
# 読み込み
with open(save_dir + 'b4_stan.pickle', mode="rb") as f:
    fit = pickle.load(f)
```

# プロット

```python
#x軸
X = df.week
X_all = df_all.week

#結果を抽出
ms = fit.extract() 

#mu
# 事後平均
mu_mean = ms['mu'].mean(axis=0)
# #95パーセンタイル抽出
mu_5 = np.array(pd.DataFrame(ms['mu']).apply(lambda x: np.percentile(x, 5), axis=0))
mu_95 = np.array(pd.DataFrame(ms['mu']).apply(lambda x: np.percentile(x, 95), axis=0))

# alpha
# 事後平均
alpha_mean = ms['alpha'].mean(axis=0)
# #95パーセンタイル抽出
alpha_5 = np.array(pd.DataFrame(ms['alpha']).apply(lambda x: np.percentile(x, 5), axis=0))
alpha_95 = np.array(pd.DataFrame(ms['alpha']).apply(lambda x: np.percentile(x, 95), axis=0))

# b_ev_len
# 事後平均
b_ev_len_mean = ms['b_ev_len'].mean(axis=0)
# #95パーセンタイル抽出
b_ev_len_5 = np.array(pd.DataFrame(ms['b_ev_len']).apply(lambda x: np.percentile(x, 5), axis=0))
b_ev_len_95 = np.array(pd.DataFrame(ms['b_ev_len']).apply(lambda x: np.percentile(x, 95), axis=0))

# b_ev_lenの効果を算出
len_effects = b_ev_len_mean*df['期間']
len_effects_5 = b_ev_len_5*df['期間']
len_effects_95 = b_ev_len_95*df['期間']
```

```python
# alpha
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X, df['2001位'], label='observed')
ax.plot(X, alpha_mean, label='predicted', c='red')
plt.fill_between(X, alpha_5, alpha_95, color='red', alpha=0.2)
plt.show()
```

```python
# mu
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X, df['2001位'], label='observed')
ax.plot(X_all, mu_mean, label='trend', c='red')
plt.fill_between(X_all, mu_5, mu_95, color='red', alpha=0.2)
# plt.legend(loc='upper left', borderaxespad=0, fontsize=15)
plt.show()
```

```python
# b_ev_len
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X, b_ev_len_mean, c='red')
plt.fill_between(X, b_ev_len_5, b_ev_len_95, color='red', alpha=0.2)
plt.show()
```

```python
# b_ev_lenの効果を算出
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X, df['期間'])
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X, len_effects, c='red')
plt.fill_between(X, len_effects_5, len_effects_95, color='red', alpha=0.2)
plt.show()
```
