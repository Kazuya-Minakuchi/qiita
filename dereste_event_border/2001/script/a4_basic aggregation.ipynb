## インポート

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font=['IPAMincho'])
```

## イベントデータ読み込み

```python
input_dir = '../event_data/'
```

```python
df = pd.read_pickle(input_dir + 'a3_output_df.pkl')
df
```

## 基礎集計・可視化

### 形式

```python
# 形式別
# --Carnival--
# データ3つのみ
# - 2001位: 平均値が他2つの2倍くらい
# - 期間: データなし
# --Groove--
# - 2001位: 全体的に一番低め。中央値 < 平均値。
# - 期間: 半分以上は174。
# --アタポン--
# - 2001位: 3つの形式では真ん中。中央値 < 平均値。
# - 期間: minはGrooveより小さく、maxはGrooveより大きい
df.groupby('形式').describe().T
```

```python
# 形式vs2001位ボーダー
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.set(font='Yu Gothic')
sns.boxplot(x='形式', y='2001位', data=df, ax=ax)
```

```python
# 形式vs期間
# - Carnivalは期間データなし
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.set(font='Yu Gothic')
sns.boxplot(x='形式', y='期間', data=df, ax=ax)
```

### 属性

```python
# 属性別
# 属性があるのはGrooveのみ
# --Da--
# - 2001位: min, mid, maxすべて一番小さい
# --Vi--
# - 2001位: 全体的に一番大きい。75%値だけ一番小さい
# --Vo--
# - 2001位: 中間くらい。25%値は一番小さい。75%値は一番大きい
df.groupby('属性').describe().T
```

```python
# 属性vs2001位ボーダー
# 属性があるのはGrooveのみ
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.set(font='Yu Gothic')
sns.boxplot(x='属性', y='2001位', data=df[df['形式']=='Groove'], ax=ax)
```

### 期間

```python
# 数値データの相関行列
# 相関はなさそう
df.corr()
```

```python
# 期間vs2001位ボーダー
# 期間が長いほどボーダーが高いかと思ったが、そうでもない
fig = plt.figure()
ax = fig.add_subplot(111)
x = df.loc[:, '期間']
y = df.loc[:, '2001位']
ax.scatter(x, y)
plt.show()
```

```python
# 期間vs2001位ボーダー
# 形式ごとに色分けしてプロット
events = df['形式'].unique()
fig = plt.figure()
ax = fig.add_subplot(111)

for event in events:
    x = df.loc[df['形式'] == event, '期間']
    y = df.loc[df['形式'] == event, '2001位']
    ax.scatter(x, y, label=event, alpha=0.5)
    
ax.legend(prop={"family":"MS Gothic"})
plt.show()
```

### 日付

```python
# 日付vs2001位ボーダー
# だんだん上がっているように見える
fig = plt.figure()
ax = fig.add_subplot(111)
x = df.sort_values('日付').loc[:, '日付']
y = df.sort_values('日付').loc[:, '2001位']
# ax.scatter(x, y)
ax.plot(x, y)
plt.show()
```

```python
# 日付vs2001位ボーダー
# 形式ごとに色分けしてプロット
# アタポンのばらつきが大きい
events = df['形式'].unique()
fig = plt.figure()
ax = fig.add_subplot(111)

for event in events:
    x = df.loc[df['形式'] == event, '日付']
    y = df.loc[df['形式'] == event, '2001位']
    # ax.scatter(x, y, label=event)
    ax.plot(x, y, label=event)
    
ax.legend(prop={"family":"MS Gothic"})
plt.show()
```

```python
# 日付vs期間
# 2018年以降は200以下
fig = plt.figure()
ax = fig.add_subplot(111)
x = df.loc[:, '日付']
y = df.loc[:, '期間']
ax.scatter(x, y)
plt.show()
```

## メモ

* 形式
  * Carnivalはデータ数少ないから分析は厳しい
  * まずはデータ数が一番多いアタポンから手を付けるのが良さそう
* 属性
  * 大きな傾向の違いはなさそう
  * 属性があるのはGrooveのみ
* 期間
  * 期間が長いほどボーダーが高いかと思ったが、そうでもなかった
  * 関係性ゼロではないと思うが、他の要因のほうが強いのかも
* 日付
  * 傾向として、だんだん上がってるように見える
  * 思いつく仮説は以下
    * プレイヤーが増えて争いが激しくなってる
    * イベント効率が上がっている
      * キャラ性能の向上
      * 楽曲、モード（グランドライブなど）の追加
* その他考えられる要因
  * イベント毎のバラつき
    * イベント報酬
  * ゲーム内の変化
    * キャラが増えてきて、放置編成しやすくなったあ
    * グランドライブができて、時間効率が上がった
